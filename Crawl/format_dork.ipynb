```python
keywords_jurnal = keywords

inurl_domains = [
    "ac.id",
    "go.id"
]

file_types = [
    "pdf",
    "docx"
]
```

```python
# UNTUK JURNAL/DOKUMEN
dorks_jurnal = []

inurl_query = " OR ".join([f"inurl:{d}" for d in inurl_domains])
filetype_query = " OR ".join([f"filetype:{ft}" for ft in file_types])

# Include sites_jurnal in the query
cleaned_journal_sites = []
for s in sites_jurnal:
    cleaned_s = s.replace('https://', '').replace('http://', '').rstrip('/')
    cleaned_journal_sites.append(f"site:{cleaned_s}")
site_journal_query = " OR ".join(cleaned_journal_sites)

for keyword in keywords_jurnal:
    dork = f'{keyword} ({inurl_query} OR {site_journal_query}) ({filetype_query})'
    dorks_jurnal.append(dork)
```

```python
for d in dorks_jurnal:
    print(d)
```

```python
dorks = []

for keyword in keywords:
    # Clean the site URL before adding to the query
    for s in sites:
        cleaned_s = s.replace('https://', '').replace('http://', '').rstrip('/')
        dork = f'{keyword} site:{cleaned_s}'
        dorks.append(dork)
```
